# 导入库

```python
import pandas as pd 
```

# 导入数据

```python
df1 = pd.read_csv('fire_archive_M6_96619.csv')
df1.head()
```

```python
df2 = pd.read_csv('fire_archive_V1_96617.csv')
df2.head()
```

```python
df3 = pd.read_csv('fire_nrt_M6_96619.csv')
df3.head()
```

```python
df4 = pd.read_csv('fire_nrt_V1_96617.csv')
df4.head()
```

# 表合并

```python
# df3,df4 合并
df5 = pd.concat([df3,df4])
df5
```

```python
# df1,df2 合并
df6 = pd.concat([df1,df2])
df6
```

```python
# 只选取两张表的 acq_date latitude	longitude
df5 = df5[['acq_date','latitude','longitude','frp']]
df6 = df6[['acq_date','latitude','longitude','frp']]
```

```python
# 合并 df5，df6，并按照时间排序
df = pd.concat([df5,df6])
df = df.sort_values(by=['acq_date'])
df
```

# 保存数据

```python
df.to_csv(path_or_buf='data.csv',index=False)
```
